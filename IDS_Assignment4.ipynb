```python

le = LabelEncoder()
b = le.fit_transform(df['beard'])
b
```

```python

h = le.fit_transform(df['hair_length'])
h
```

```python
s = le.fit_transform(df['scarf'])
s
```

```python
e = le.fit_transform(df['eye_color'])
e
```

```python
data = df.drop("beard", axis= 'columns')
data = df.drop("hair_length", axis= 'columns')
data = df.drop("scarf", axis= 'columns')
data = df.drop("eye_color", axis= 'columns')
data["beard"] = b
data["hair_length"] = h
data["scarf"] = s
data["eye_color"] = e
data
     
```

```python

#Here only excluding the 'gender' and taking the remaining as inputs
X = data.drop("gender", axis= 'columns')
#Only taking 'gender' as the output
Y = data['gender']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.20, random_state=7)
```

```python

#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(X_train,y_train)
predictions = rf.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score, f1_score
#print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
#Support Vector Machine Classifier
from sklearn.svm import SVC
svm= SVC()
svm.fit(X_train,y_train)
predictions = svm.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score, f1_score
#print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python

#MLP Classifier
from sklearn.neural_network import MLPClassifier
mlp= MLPClassifier()
mlp.fit(X_train,y_train)
predictions = mlp.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score, f1_score
#print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
#Decision Tree Classifier with Monte Carlo (shuffle-split) Cross Validation
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
from sklearn.model_selection import ShuffleSplit,cross_val_score
shuffle_split=ShuffleSplit(test_size=0.33,n_splits=5)
scores=cross_val_score(dt,X_test,predictions,cv=shuffle_split)
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score, f1_score
print(classification_report(y_test,predictions))
#print(confusion_matrix(y_test,predictions))
#print("Average Cross Validation score :{}".format(scores.mean()))
```
